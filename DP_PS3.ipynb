```python
import pandas as pd
user_cols=['user_id','age','gender','occupation','zip_code']
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user',sep='|', header =None,skiprows=1,names=user_cols)


#users.head() -> prints first 5 rows.
#print(type(users))
#print(type(users['user_id']))
#html

#Q1_Step 4: group the tables by occupation and find the mean value for each age.
mean_group= users[['occupation', 'age']].groupby('occupation')
mean_value=mean_group.mean()
print(mean_value)

```

```python
#Q1_STEP 5: Find male ratio per occupation to the overall population per occupation.

data = users[['gender', 'occupation']]

# Now, filter 'male' from gender in 'data'
males = data.loc[data['gender'] == 'M']

# Group by Occupation and count in df
group_by_occupations = data.groupby('occupation').count()

# Group by Occupation and count in males
male_count_by_occupation = males.groupby('occupation').count()

# male_ratio_per_occupation
male_ratio_per_occupation = male_count_by_occupation / group_by_occupations

male_ratio_per_occupation.sort_values(ascending=False, by='gender')
```

```python
#Q1_Step6: . For each occupation, calculate the minimum and maximum ages

occ_min_age = mean_group.min()

occ_max_age = mean_group.max()

result = pd.concat([occ_max_age, occ_min_age], axis=1)

result
```

```python
#Q1_Step7: For each combination of occupation and sex, calculate the mean age

gen_occ_age = users[['occupation', 'gender', 'age']]

users_mean_age_by_occupation_and_gender = gen_occ_age.groupby(['occupation', 'gender']).mean('age')

users_mean_age_by_occupation_and_gender
```

```python
#Q1_Step 8: For each occupation present the percentage of women and men.

gender_percentage_by_occupation = data.groupby('occupation')['gender'].value_counts(normalize=True)*100

gender_percentage_by_occupation
```

```python
import pandas as pd
#Step2
data = pd.read_table('wind.txt',delimiter='\s+')

#data.head()
#Step3
data.rename(columns={"Yr":"Year", "Mo":"Month", "Dy":"Day"}, inplace=True)

data
```

```python
#Step4
def fixYear():
    data['Year'] += 1900 #

fixYear() # Fix year by converting it in 4-digit value

data
```

```python
#Step5
data.index = pd.to_datetime(pd.concat([data['Year'], data['Month'], data['Day']], axis=1)) #setting index

data.drop(['Year','Month','Day'], axis=1, inplace=True) 

data

```

```python
data.index # datatype is datetime64
```

```python
#Q4_Step6:

data.isna().sum() # Null values in each column


data.isnull().sum().sum() # Sum of all null values 
```

```python
#Q4_Step7:
data.notnull().sum().sum() # Counts non-missing values
```

```python
#Step8:
data.mean().mean() # mean of windspeeds for all locations and times.
```

```python
#Step9:
# New dataframe having 4 columns: Min, Max, Mean & Std Dev
loc_stats = pd.concat([data.min(), data.max(), data.mean(), data.std()], axis=1) 

# Rename the columns with proper labels
loc_stats.rename(columns={0:'Min', 1:'Max', 2:'Mean', 3:'Std'}, inplace=True) 

loc_stats
```

```python
#Step_10:
# New daily_stats dataframe having 4 columns
day_stats = pd.concat([data.min(axis=1), data.max(axis=1), data.mean(axis=1), data.std(axis=1)], axis=1)

# Rename the columns
day_stats.rename(columns={0:'Min',1:'Max',2:'Mean',3:'Std'}, inplace=True)

day_stats
```

```python
#Step_11
avg_windspeed_in_jan = data[data.index.month == 1].mean()

avg_windspeed_in_jan
```

```python
#Step_12
yearly = data.resample('Y').ffill()

yearly
```

```python
#Step_13
monthly = data.resample('M').ffill()

monthly
```

```python
#Step14
weekly = data.resample('W').ffill()

weekly
```

```python
weekly_stats = weekly[:52]

weekly_stats.shape[0] 
```

```python
#Step_15
weekly_stats = pd.concat([weekly_stats.mean(), weekly_stats.max(), weekly_stats.mean(), weekly_stats.std()], axis=1)
weekly_stats.rename(columns={0:'Mean',1:'Max',2:'Mean',3:'Std'},inplace=True)

weekly_stats
```

```python


import pandas as pd
import matplotlib.pyplot as plt

data1 = pd.read_csv('us-marriages-divorces-1867-2014.csv')

#data1

fig = plt.figure() # an empty figure with no axes

plt.rcParams["figure.figsize"] = [15, 10] # set the figure size

# Set x & y axis and legends
data1.plot( 'Year', ['Marriages_per_1000','Divorces_per_1000'], label=['Marriages', 'Divorces'])

# Set Title
plt.title('Marriages V/S Divorces per capita in the U.S. between 1867 and 2014', color = 'k', fontsize = 22)

# Set labels
plt.xlabel('Year', color = 'k', fontsize = 17)
plt.ylabel('Capita', color = 'k', fontsize = 17)

plt.show()
```

```python
#Question_7
filter_by_years = (data1['Year'] == 1900) | (data1['Year'] == 1950) | (data1['Year'] == 2000) 

data_table = data1.loc[filter_by_years, ['Year','Marriages_per_1000','Divorces_per_1000']]

data_table
```

```python
data_table[{'Marriages_per_1000','Divorces_per_1000'}].plot.bar()

plt.title("Marriages and Divorces per capita in the U.S. between 1900, 1950 and 2000", color = 'k', fontsize = 20)

plt.xlabel('Year', color = 'k', fontsize = 15)
plt.ylabel('Capita', color = 'k', fontsize = 15)

plt.xticks((0, 1, 2), ("1900", "1950", "2000"), rotation = 0)
```
